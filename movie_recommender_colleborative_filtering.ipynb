```python
import ast

def convert(obj):
    """
    Extracts the 'name' values from a list of dictionaries or a list of strings.

    This function first checks if the input `obj` is a string. If it is, it assumes
    it's a string representation of a list of dictionaries and uses `ast.literal_eval`
    to convert it into a Python list. Then, it extracts the 'name' values from the
    dictionaries within the list.

    If `obj` is not a string, it assumes it's already a list of strings and simply
    returns it as is.

    Args:
        obj (str or list): The input object, which can be a string representation
                           of a list of dictionaries or a list of strings.

    Returns:
        list: A list of strings representing the 'name' values extracted from
              the input object.
    """
    if isinstance(obj, str):
        L = []
        for i in ast.literal_eval(obj):
            L.append(i['name'])
        return L
    else:  # If obj is not a string, assume it's already a list of strings
        return obj

movies['genres'] = movies['genres'].apply(convert)
```

```python
import pandas as pd
import ast
import numpy as np

# Assuming your 'movies' DataFrame is already created as shown in your notebook

def combine_lists(row):
    genres = " ".join(row['genres']) if isinstance(row['genres'], list) else str(row['genres']) if not pd.isna(row['genres']) else ""
    keywords = " ".join(row['keywords']) if isinstance(row['keywords'], list) else str(row['keywords']) if not pd.isna(row['keywords']) else ""
    cast_list = ast.literal_eval(row['cast']) if isinstance(row['cast'], str) else []
    cast = " ".join([actor['name'] for actor in cast_list])
    crew_list = ast.literal_eval(row['crew']) if isinstance(row['crew'], str) else []
    director = ""
    for member in crew_list:
        if member['job'] == 'Director':
            director = member['name']
            break
    overview = " ".join(row['overview']) if isinstance(row['overview'], list) else str(row['overview']) if not pd.isna(row['overview']) else ""

    return f"{genres} {keywords} {cast} {director} {overview}"

# Apply the function to create the 'tags' column in the 'movies' DataFrame
movies['tags'] = movies.apply(combine_lists, axis=1)

# Create a new DataFrame 'new_df' with a copy of the 'tags' column and lowercase it
new_df = movies[['tags']].copy()
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

# Display the head of the new_df to check the result
print(new_df.head())
```

So yeah, in NLP (natural language processing), we got *lots of methods* to convert text into numbers (vectors), coz machine dont understand words like we do. Here's all the major methods we use for converting text to vectors:

---

### 1. **Bag of Words (BoW)**
- This is the basic one.
- It just count how many times each word appear in the text.
- Doesn't care about order of words, just presence and frequency.
- Example:  
  - I like cats  [1, 1, 1]  
  - I like like dogs  [2, 0, 1]

 **Pros**: simple and fast  
 **Cons**: no context, huge vectors if vocab is big

---

### 2. **TF-IDF (Term Frequency - Inverse Document Frequency)**
- Like BoW but smarter.
- Downweights common words like the, is, etc.
- Gives more importance to rare but useful words.

 **Pros**: better than BoW for relevance  
 **Cons**: still no word order or context

---

### 3. **Word2Vec**
- Trained on huge text data to learn word meanings.
- Words are represented as dense vectors, where similar meaning words are close in space.
- Two types: CBOW (predict word from context) and Skip-gram (predict context from word)

 **Pros**: captures semantic meaning  
 **Cons**: fixed word vector, can't handle out-of-vocab words

---

### 4. **GloVe (Global Vectors)**
- Like Word2Vec but use global word co-occurrence matrix.
- Still gives vector for each word.
- Pretrained embeddings available.

 **Pros**: fast and meaningful vectors  
 **Cons**: same problem like Word2Vec  no context of sentence

---

### 5. **FastText**
- Facebook made it.
- Improves on Word2Vec by using subword info (character n-grams).
- Can handle unknown words better.

 **Pros**: better at rare or new words  
 **Cons**: more computationally heavy

---

### 6. **Doc2Vec**
- Like Word2Vec but for whole sentences or documents.
- Gives one vector per document, not per word.

 **Pros**: can represent whole doc meaning  
 **Cons**: hard to train, not as popular now

---

### 7. **Transformer-based Embeddings (like BERT, RoBERTa, GPT etc.)**
- These are state-of-the-art now.
- They look at whole sentence before deciding meaning of word (context-aware).
- Output vector for each word and whole sentence.

 **Pros**: crazy good, understands grammar, sarcasm, context  
 **Cons**: needs lot of compute, slower than others

---

### 8. **Sentence Embeddings (e.g., Sentence-BERT, Universal Sentence Encoder)**
- These give one vector for full sentence.
- Useful for similarity, search, classification.

 **Pros**: easy to plug and play for many tasks  
 **Cons**: little slower and models are big

---

### Summary Table (for vibes ):

| Method         | Context-Aware | Handles OOV | Vector Type     |
|----------------|----------------|--------------|------------------|
| BoW            |               |             | Sparse           |
| TF-IDF         |               |             | Sparse           |
| Word2Vec       |               |             | Dense (word)     |
| GloVe          |               |             | Dense (word)     |
| FastText       |               |             | Dense (word)     |
| Doc2Vec        |               |             | Dense (doc)      |
| BERT etc.      |               |             | Dense (context)  |
| Sentence-BERT  |               |             | Dense (sentence) |

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

cv = CountVectorizer(max_features=5000,stop_words='english')

# Fit and transform the 'tags' column and store the result in 'vector'
vector = cv.fit_transform(new_df['tags']).toarray()

# Calculate cosine similarity using 'vector'
similarity = cosine_similarity(vector)
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = similarity[index]  # Assign similarity values to distances
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1]) [1:6] # Get the sorted list of movie indices and distances
                                                                              # Select the top 5 similar movies (excluding the movie itself)
    for i in movies_list:
      print(new.iloc[i[0]].title) # Print the titles of recommended movies
```
